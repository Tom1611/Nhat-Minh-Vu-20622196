```python
## 1.1 Initial Prompt & Pseudocode (8 marks)

Course Reference:
I used the planning method from Week 2 - Activity 3: User Preferences with Simplified Methodology.

## My Prompt to AI:

i wanna create pseudocode for a CLI Task Manager program using the planning methodology from my introductory Python course.

Requirements:
- The program should allow users to add, list, and remove tasks stored in memory
_ Basic lists and dictionaries for data storage
_ Simple functions with parameters
_ Basic file operations (if needed for persistence)
_ While loops for menu systems
_ Try/except for basic error handling only

Please provide a structured pseudocode outline that includes:
1. State the Problem
2. Describe Input and Output
3. Work the Problem by Hand
4. Develop an Algorithm (Pseudocode)
5. Test Plan


## AI responds:


1. State the Problem

Create a simple Command-Line Interface (CLI) Task Manager program that allows users to:

Add tasks

List all tasks

Remove tasks

The tasks are stored in memory using Python data structures (lists/dictionaries). Optionally, tasks can be saved and loaded from a file to maintain persistence between sessions. The menu should loop until the user chooses to exit.

2. Describe Input and Output
Input:

User choices from a text-based menu:

Add a new task (description entered as a string)

List all current tasks

Remove a task (by number or name)

Exit the program

(Optional) Confirmation prompts

Output:

Display of menu options

Messages indicating success/failure of actions

List of tasks with numbering

(Optional) Task persistence via reading/writing from a text file

3. Work the Problem by Hand

Example flow:

User runs the program

1. Menu appears:
   [1] Add Task
   [2] List Tasks
   [3] Remove Task
   [4] Exit

2. User selects 1:
   - Enters task: "Buy groceries"
   - Task added

3. User selects 1 again:
   - Enters task: "Call Alice"
   - Task added

4. User selects 2:
   - Program displays:
     1. Buy groceries
     2. Call Alice

5. User selects 3:
   - Enters task number to remove: 1
   - "Buy groceries" removed

6. User selects 4:
   - Program exits

4. Develop an Algorithm (Pseudocode)
START PROGRAM

DEFINE task_list AS empty list

FUNCTION show_menu()
    PRINT "[1] Add Task"
    PRINT "[2] List Tasks"
    PRINT "[3] Remove Task"
    PRINT "[4] Exit"

FUNCTION add_task(task_list)
    PROMPT "Enter task description:"
    READ task_description
    APPEND task_description TO task_list
    PRINT "Task added."

FUNCTION list_tasks(task_list)
    IF task_list is empty
        PRINT "No tasks to show."
    ELSE
        FOR EACH task IN task_list WITH index i
            PRINT i+1, task

FUNCTION remove_task(task_list)
    CALL list_tasks(task_list)
    IF task_list is NOT empty
        PROMPT "Enter task number to remove:"
        TRY
            READ task_number
            CONVERT task_number TO integer
            IF task_number is within valid range
                REMOVE task at index task_number - 1 FROM task_list
                PRINT "Task removed."
            ELSE
                PRINT "Invalid task number."
        EXCEPT ValueError
            PRINT "Please enter a valid number."

MAIN LOOP:
    WHILE True
        CALL show_menu()
        PROMPT "Choose an option (1-4):"
        TRY
            READ choice
            CONVERT choice TO integer
            IF choice == 1
                CALL add_task(task_list)
            ELSE IF choice == 2
                CALL list_tasks(task_list)
            ELSE IF choice == 3
                CALL remove_task(task_list)
            ELSE IF choice == 4
                PRINT "Goodbye!"
                BREAK
            ELSE
                PRINT "Invalid choice."
        EXCEPT ValueError
            PRINT "Please enter a number."

END PROGRAM


✅ Optional enhancement for persistence:

On program start, read tasks from a file if it exists

On each change (add/remove), write tasks back to the file

5. Test Plan
Test Case	Input	Expected Output
Add Task	"Do homework"	Task added, shown in task list
List Tasks (non-empty)	N/A	Display list of tasks with numbers
Remove Valid Task	Enter number "1"	Task removed, updated list shown
Remove Invalid Task (number too big)	Enter number "5"	"Invalid task number" message
Remove Task (non-integer)	Enter "abc"	"Please enter a valid number" message
Exit Program	Enter "4"	Program prints "Goodbye!" and exits
Invalid Menu Input	Enter "9" or "abc"	"Invalid choice" or error handled
List Tasks (empty)	N/A	"No tasks to show."
```
